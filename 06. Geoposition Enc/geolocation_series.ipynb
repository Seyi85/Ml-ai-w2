```python
def online(x):
    if 'Digital' in x:
            return 1
    if 'Internet' in x:
        return 1
    else:
        return 0
shops['online']= shops['shop_name'].apply(online)
shops
```

```python
def split(x):
    xy = x.split(" ")
    xy[0].strip()
    if xy[0]=="!" or xy[0]=="":
        if xy[1] == "St.":
            rename = "St. " + xy[2]
            return rename
        return xy[1]
    else:
        return xy[0]
shops['shop_city']= shops['shop_name'].apply(split)
shops   
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent='Paul')
#location = geolocator.geocode(city+','+ country)
```

```python
shops['longitude']= shops['shop_city'].apply(lambda x:geolocator.geocode(x).longitude if geolocator.geocode(x) else None)
```

```python
shops['latitude']= shops['shop_city'].apply(lambda x:geolocator.geocode(x).latitude if geolocator.geocode(x) else None)
```
