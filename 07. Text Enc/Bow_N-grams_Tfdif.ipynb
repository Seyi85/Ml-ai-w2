```python
import pandas as pd
import numpy as np
#import spacy
import time
from tqdm import tqdm_notebook as tqdm
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df_items = pd.read_csv('items.csv')
df_cat= pd.read_csv('categories.csv')
```

```python
df_cat
```

```python
#to select the categories
cat_name = df_cat['category_name']
cat_name    
```

```python
cat_name.apply(lambda x: x.split("-")[0])
```

```python
item_name = df_items['item_name']
item_name
```

```python
bow = CountVectorizer(min_df =0.1)
item_name_bow = bow.fit_transform(item_name)
tokens = bow.get_feature_names()
pd.DataFrame(data=item_name_bow.toarray(),columns=tokens)
```

```python
bow = CountVectorizer(ngram_range=(1, 2), min_df=0.01)
item_name_bow = bow.fit_transform(item_name)
tokens = bow.get_feature_names()
pd.DataFrame(data=item_name_bow.toarray(),columns=tokens)
```

```python
bow = CountVectorizer(ngram_range=(2, 2), min_df=0.01)
item_name_bow = bow.fit_transform(item_name)
tokens = bow.get_feature_names()
pd.DataFrame(data=item_name_bow.toarray(),columns=tokens)
```

```python
tfidf = TfidfVectorizer(min_df = 0.01)
item_name_tfidf = tfidf.fit_transform(item_name)
tokens = tfidf.get_feature_names()
pd.DataFrame(data=item_name_tfidf.toarray(),columns=tokens)
```

```python
tfidf = TfidfVectorizer(ngram_range=(1, 2), min_df=0.01)
item_name_tfidf = tfidf.fit_transform(item_name)
tokens = tfidf.get_feature_names()
pd.DataFrame(data=item_name_tfidf.toarray(),columns=tokens)
```

```python
tfidf = TfidfVectorizer(ngram_range=(2, 2), min_df=0.01)
item_name_tfidf = tfidf.fit_transform(item_name)
tokens = tfidf.get_feature_names()
pd.DataFrame(data=item_name_tfidf.toarray(),columns=tokens)
```
